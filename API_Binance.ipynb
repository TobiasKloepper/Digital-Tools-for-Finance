`````python
# Here is some Python Code that tries to apply that:
import requests
import datetime
import json
import pandas as pd
import os
import numpy

ROOT_URL = "https://api.binance.com"
endpoint = "api/v3/klines"
symbol = "BTCUSDT"
interval = "1d"
startTime = "20210615"
limit = "75"

request_url = (ROOT_URL + "/{endpoint}?symbol={symbol}&interval={interval}&startTime={startTime}&limit={limit}").format(endpoint = endpoint,
                                                                                                                      symbol = symbol,
                                                                                                                      interval = interval,
                                                                                                                      startTime = startTime,
                                                                                                                      limit = limit)

request_url
`````

`````python
# Send Request and ask for response:

response = requests.get(request_url)
response
`````
