# The Student's t-distribution

Here is some terribly inefficient code that plots the density of a t-distribution with interactive slider!

```python
import numpy as np
import math as maths
import matplotlib.pyplot as plt

def students_t(t, df):
    pdf = (maths.gamma((df + 1) / 2) / (np.sqrt(df * np.pi) * maths.gamma(df / 2))) * (1 + (t ** 2)/df) ** (-(df + 1)/2)
    
    return pdf
```

```python
def plot_t(df):

    pdf = np.zeros(1000)
    t = np.linspace(-10, 10, 1000)
    i = 1;

    for k in range(0, 1000):
        pdf[k] = students_t(t[k], df)
    
    plt.plot(t, pdf);
    plt.ylim(top = 0.4)

```

```python
from ipywidgets import interact
interact(plot_t, df = (0.5, 10));
```

## Bonus!
This is all of them plotted at the same time!

```python
df = 1
pdf = np.zeros((14, 1000))
t = np.linspace(-10, 10, 1000)
i = 1;
while i < 15:
    for k in range(0, 1000):
        pdf[i - 1, k] = students_t(t[k], i)
    i += 1

plt.plot(t, np.transpose(pdf));
```
